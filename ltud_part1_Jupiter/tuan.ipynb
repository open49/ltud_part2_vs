```python
from PIL import Image, ImageOps
Im = Image.open('0_buildings_-_lambs_house_002475_underexposed1024.jpg') 
Im = ImageOps.grayscale(Im)

from matplotlib import pyplot as plt
Im_eq = ImageOps.equalize(Im)

fig,axs = plt.subplots(1,2,figsize=(10,5))
axs[0].imshow(Im, cmap = 'gray')
axs[0].set_title('Ảnh gốc')
axs[1].imshow(Im_eq, cmap ='gray')
axs[1].set_title('Ảnh sau khi chuẩn hóa histogram')


plt.show()
```
